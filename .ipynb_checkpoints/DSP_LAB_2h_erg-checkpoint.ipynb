Definition of Functions

```python
def T_q(i):   #calculating absolute threshold of hearing
    f = i*fs/N;
    return 3.64*((f/1000)**(-0.8)) - 6.5*np.exp(-0.6*((f/1000 - 3.3)**2)) + (10**(-3))*((f/1000)**4)


def St(k, power_spectrum): #locating the tone masks
    bool = True 
    i = 2
    if k <= 2:
        bool = False
    elif 2 < k < 63 :
        predicate_1 = power_spectrum[k] > power_spectrum[k - 1] and power_spectrum[k] > power_spectrum[k + 1]
        bool = predicate_1 and bool
        while i <= 2:
            predicate = power_spectrum[k] > (power_spectrum[k - i] + 7) and power_spectrum[k] > (power_spectrum[k + i] + 7)
            bool = predicate and bool
            i = i + 1
    elif 63 <= k < 127 :
        predicate_1 = power_spectrum[k] > power_spectrum[k - 1] and power_spectrum[k] > power_spectrum[k + 1]
        bool = predicate_1 and bool
        while i <= 3:
            predicate = power_spectrum[k] > (power_spectrum[k - i] + 7) and power_spectrum[k] > (power_spectrum[k + i] + 7)
            bool = predicate and bool
            i = i + 1
    elif 127 <= k <= 250 :
        predicate_1 = power_spectrum[k] > power_spectrum[k - 1] and power_spectrum[k] > power_spectrum[k + 1]
        bool = predicate_1 and bool
        while i <= 6:
            predicate = power_spectrum[k] > (power_spectrum[k - i] + 7) and power_spectrum[k] > (power_spectrum[k + i] + 7)
            bool = predicate and bool
            i = i + 1
    return bool


def SF(i, j, half_frame): #calculating masks' coverage area
    f_i = i*fs/N;
    f_j = j*fs/N;
    b_i = 13*np.arctan(0.00076*f_i) + 3.5*np.arctan((f_i/7500)**2) # bark scale
    b_j = 13*np.arctan(0.00076*f_j) + 3.5*np.arctan((f_j/7500)**2) # bark scale
    delta = b_i - b_j  #we assume (b_i - b_j) belongs to the set [-3,8)
    if(delta>=-3 and delta<-1): return (17*delta - 0.4*half_frame[j] + 11)
    elif(delta<0): return ((0.4*half_frame[j] + 6)*delta)
    elif(delta<1): return -17*delta
    else: return ((0.15*half_frame[j] - 17)*delta - 0.15*half_frame[j])

    
def T_tm(i,j, half_frame): #calculating individual masking thresholds (tone masks)
    f_i = i*fs/N;
    f_j = j*fs/N;
    b_i = 13* np.arctan(0.00076*f_i) + 3.5*np.arctan((f_i/7500)**2) # bark scale
    b_j = 13* np.arctan(0.00076*f_j) + 3.5*np.arctan((f_j/7500)**2) # bark scale
    delta = b_i - b_j
    if(delta<-3 or delta>=8): return 0
    else: return (half_frame[j] - 0.175*b_j + SF(i,j,half_frame) - 6.025)

    
def T_nm(i,j, half_frame): #calculating individual masking thresholds (noise masks)
    f_i = i*fs/N;
    f_j = j*fs/N;
    b_i = 13* np.arctan(0.00076*f_i) + 3.5*np.arctan((f_i/7500)**2) # bark scale
    b_j = 13* np.arctan(0.00076*f_j) + 3.5*np.arctan((f_j/7500)**2) # bark scale
    delta = b_i - b_j
    if(delta<-3 or delta>=8): return 0
    else: return (half_frame[j] - 0.175*b_j + SF(i,j,half_frame) - 2.025)
    
def T_g(i, pinakas): #calculating global masking thresholds
    sum_tone = 0
    sum_noise = 0
    for j in range(256):
        if((isinstance(pinakas[i][j], int)==0)):
            T_noise = pinakas[i][j][0]
            T_tone = pinakas[i][j][1]
            if(T_noise>0): sum_noise += 10**(0.1*T_noise)
            if(T_tone>0): sum_tone += 10**(0.1*T_tone)
    return 10*np.log10(10**(0.1*T_q(i)) + sum_noise + sum_tone)
```

Extra function: We use the following function so that specimens with zero values are not shown in stem plot
(found in: https://stackoverflow.com/questions/18697417/not-plotting-zero-in-matplotlib-or-change-zero-to-none-python )

```python
def zero_to_nan(values):
    #Replace every 0 with 'nan' and return a copy.
    return [float('nan') if x==0 else x for x in values]
```

Step 1.0

```python
music_stereo, f1 = sf.read('music.wav')
music_mono = music_stereo[:,0]/2 + music_stereo[:,1]/2 # Convert to mono format

music = music_mono / np.amax(abs(music_mono)) # Music normalized 

MAX = np.amax(abs(music_mono))
print(MAX)

N = 512    #frame length
fs = 44100 #sampling frequency
hann = np.hanning(N)
                             
size = np.size(music)
frames = []                             

i = 0 
while i <= (size - 512):
    frames.append(music[i:i+512]*hann)   #List of frames of 512 specimens windowed with hanning
    i = i + 512

Zeros = np.zeros(512 - (size - i)) #Zero-pad Last Frame to have 512 specimens 
frames.append(np.concatenate((music[i:], Zeros), axis = None)*hann)
```

Step 1.1

```python
%matplotlib inline
freq = np.arange(0,20000,1)
barks = 13* np.arctan(0.00076*freq) + 3.5*np.arctan((freq/7500)**2) # bark scale

#defining bark scale for plotting frequency bins in the set [1,255] (in some cases freq. bin i=0 should be excluded)

i_without_0 = np.arange(1,256,1)
f_without_0 = i_without_0*fs/N
b_without_0 = 13* np.arctan(0.00076*f_without_0) + 3.5*np.arctan((f_without_0/7500)**2) # bark scale

#defining bark scale for plotting frequency bins in the set [0,255]

i = np.arange(0,256,1)
f = i*fs/N
b = 13* np.arctan(0.00076*f) + 3.5*np.arctan((f/7500)**2)

plt.figure(1)
plt.plot(freq,barks)
plt.xlabel('Center Frequency in Hz')
plt.ylabel('Bark')
plt.title("Frequency-Bark Relation")

abs_thres = T_q(i_without_0)

plt.figure(2)
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Absolute Threshold of Hearing")
```

```python
#calculating the power spectrum of each frame

Power_Spectrum = []
PN = 90.302

for frame in frames:
    p = 20*np.log10(abs(np.fft.fft(frame))) + PN
    p_cut = p[:256]
    Power_Spectrum.append(p_cut)    

#plotting the first 3 power spectrums alongside the absolute threshold of hearing

plt.figure(1)
plt.plot(b,Power_Spectrum[0])
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Power Spectrum of 1st Frame")
plt.figure(2)
plt.plot(b,Power_Spectrum[1])
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Power Spectrum of 2nd Frame")
plt.figure(3)
plt.plot(b,Power_Spectrum[2])
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Power Spectrum of 3rd Frame")
```

Step 1.2

```python
#calculating tone masks

Tone_Masks = []
for power_spectrum in Power_Spectrum:
    Tone_Masks.append(np.zeros(256))
    for k in range(250):
        if(St(k, power_spectrum)):
            Tone_Masks[-1][k] = 10*np.log10(10**(0.1*power_spectrum[k-1]) + 10**(0.1*power_spectrum[k]) + 10**(0.1*power_spectrum[k+1]))

Tone_Masks = np.transpose(Tone_Masks) #we calculated the transpose matrix so that it matches the shape of given noise mask matrix
Noise_Masks = np.load('P_NM.npy')

# print(np.shape(Tone_Masks))
# print(np.shape(Noise_Masks))

```

```python
#printing the Tone & Noise Masks for random frame (e.g. 50th frame)

tone_masks = zero_to_nan(Tone_Masks[:,49])
noise_masks = zero_to_nan(Noise_Masks[:,49])

plt.plot(b,Power_Spectrum[49])

markerline_tone, stemline_tone, baseline_tone = plt.stem(b, tone_masks, markerfmt = 'kx')
markerline_noise, stemline_noise, baseline_noise = plt.stem(b, noise_masks, markerfmt = 'ro')

markerline_noise.set_markerfacecolor('none') #makes the marks unfilled

stemline_tone.remove() #removes the stem line
stemline_noise.remove()

plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Tone Masks (x) and Noise Masks (o) of 50th frame")
```

Step 1.3

```python
#loading new masks

Tone_Masks_cut = np.load('P_TMc.npy')
Noise_Masks_cut = np.load('P_NMc.npy')

#printing the (now reduced) Tone & Noise Masks for the same frame as before (50th frame)

tone_masks_cut = zero_to_nan(Tone_Masks_cut[:,49])
noise_masks_cut = zero_to_nan(Noise_Masks_cut[:,49])

plt.plot(b,Power_Spectrum[49])

markerline_tone, stemline_tone, baseline_tone = plt.stem(b, tone_masks_cut, markerfmt = 'kx')
markerline_noise, stemline_noise, baseline_noise = plt.stem(b, noise_masks_cut, markerfmt = 'ro')

markerline_noise.set_markerfacecolor('none')

stemline_tone.remove()
stemline_noise.remove()
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Tone Masks (x) and Noise Masks (o) of 50th frame (Reduced)")
```

Step 1.4

```python
#calculating Individual Masking Thresholds (this may take some time so please be patient) 

Individual_Masking_Thresholds = []
for counter in range(1179):
    half_frame_tone = Tone_Masks_cut[:,counter]
    half_frame_noise = Noise_Masks_cut[:,counter]
    j_noise = []
    j_tone = []
    Thresholds_per_half_frame = np.zeros((256,256), dtype=object) #we initialize a 2D array in which we want to store tuples
    for j in range(256):
        if(half_frame_tone[j]>0):
            j_tone.append(j)
        if(half_frame_noise[j]>0):
            j_noise.append(j)
    for j in j_noise:
        for i in range(256):
            T_noise = T_nm(i,j,half_frame_noise)
            Thresholds_per_half_frame[i][j] = (T_noise, 0) #the tuple's first element is the noise-mask threshold
    for j in j_tone:
        for i in range(256):
            T_tone = T_tm(i,j,half_frame_tone)
            if((isinstance(Thresholds_per_half_frame[i][j], int)==0)): Thresholds_per_half_frame[i][j] = (Thresholds_per_half_frame[i][j][0], T_tone)
            else: Thresholds_per_half_frame[i][j] = (0, T_tone) #the tuple's second element is the tone-mask threshold
    Individual_Masking_Thresholds.append(Thresholds_per_half_frame)
            
```

```python
#plotting individual masking thresholds (tone masks) for 50th frame

tone_masks_cut = zero_to_nan(Tone_Masks_cut[:,49])

var = -1
for j in Tone_Masks_cut[:,49]:
    var += 1                       #var represents the index of the tone mask
    if(j>0):
        lista = []
        for i in range(256): lista.append(Individual_Masking_Thresholds[49][i][var][1]) #we store the second element of the tuple: aka the tone-mask threshold
        plt.plot(b,lista)

plt.plot(b_without_0, abs_thres, 'm--')

markerline_tone, stemline_tone, baseline_tone = plt.stem(b, tone_masks_cut, markerfmt = 'kx')

stemline_tone.remove()
plt.title("Individual Masking Thresholds of 50th frame (Tone Masks)")
```

Step 1.5

```python
#calculating global masking thresholds (this may take some time so please be patient) 

Global_Masking_Thresholds = []

for counter in range(1179):
    Thresholds_list = []
    pinakas = Individual_Masking_Thresholds[counter]
    i = 1                
    while(i<256):
        total = T_g(i, pinakas)
        Thresholds_list.append(total)
        i += 1
    Global_Masking_Thresholds.append(Thresholds_list)
```

```python
#plotting global masking thresholds for 50th frame

plt.plot(b,Power_Spectrum[49])
plt.plot(b_without_0,Global_Masking_Thresholds[49])
plt.plot(b_without_0, abs_thres, 'm--')
plt.title("Global Masking Thresholds of 50th frame")
```

Part 2

Step 2.0

```python
# Definition of analysis and synthesis Filters

H_Filters = []
G_Filters = []

M = 32
L = 2*M
n = np.arange(0,L,1)
n1 = 2*M - 1 - n


for k in range(M):
    H_Filters.append(np.sin((n+1/2)*np.pi/(2*M))*np.sqrt(2/M)*np.cos((2*n+M+1)*(2*k+1)*np.pi/(4*M)))
    G_Filters.append(np.sin((n1+1/2)*np.pi/(2*M))*np.sqrt(2/M)*np.cos((2*n1+M+1)*(2*k+1)*np.pi/(4*M)))
```

Step 2.1

```python
# Convolution with analysis filters and Decimation so that signal's time components derive

#         Time_compon
#         /   | ... \   M = 32 filters
#      /...\   ....     F = 1179 Frames per filter          

time_compon = []

music = music_mono / np.amax(abs(music_mono)) # Music normalized 

N = 512 # frame length
                             
size = np.size(music)
frames = []                             

i = 0 
while i <= (size - 512):
    frames.append(music[i:i+512])   # List of frames of 512 specimens
    i = i + 512

Zeros = np.zeros(512 - (size - i)) # Zero-pad Last Frame to have 512 specimens 
frames.append(np.concatenate((music[i:], Zeros), axis = None))


for k in range(M):
    frames_processed = []
    for frame in frames:
        u = np.convolve(frame, H_Filters[k],'full')
        y = u[0:np.size(u):M]  # decimation
        frames_processed.append(y)
    time_compon.append(frames_processed)
```

Step 2.2

```python
# Finding Bits allocated to each frame of the signal

fs = 44100
N = 512

B = 16
R = 2**B
Bits = []

k = 1
while k <= 32:
    indexes = []
    fk = (2*k - 1)*fs*np.pi/(2*M)
    r = fs*np.pi/(2*M)
    i = 1
    while i < 256:
        f = i*(2*np.pi/N)*fs
        if  (fk - r <= f) and (f <= fk + r): 
            indexes.append(i - 1)             # indexes corresponding to filter passband
        i = i + 1
    b = []
    for frame in range(1179):
        Tg = min([Global_Masking_Thresholds[frame][j] for j in indexes]) # min Global masking threshold
        b.append(int(np.ceil((np.log2(R/Tg) - 1))))
    Bits.append(b)
    k = k + 1  
```

```python
# Simple Quantization function of a value (val)

def quantize(val, to_values):
    best_match = None
    best_match_diff = None
    for other_val in to_values:
        diff = abs(other_val - val)
        if best_match is None or diff < best_match_diff:
            best_match = other_val
            best_match_diff = diff
    return best_match
```

```python
# Quantization with Adjustive Quantizer (this may take some time so please be patient)

time_compon_Quant = [] 

for k in range(M):
    frames_Quant = []
    for frame in range(1179):
        Bk = Bits[k][frame]
        Delta = 2**Bk
        levels = np.linspace(np.amin(time_compon[k][frame]), np.amax(time_compon[k][frame]), Delta)
        Quant = np.array([])
        for i in time_compon[k][frame]:
            Quant = np.append(Quant,quantize(i,levels))
        frames_Quant.append(Quant)
    time_compon_Quant.append(frames_Quant)
```

```python
# Quantization with Non Adjustive Quantizer (this may take some time so please be patient)

time_compon_Quant_stable = [] 

for k in range(M):
    frames_Quant = []
    for frame in range(1179):
        Bk = 8
        Delta = 2**Bk
        levels = np.linspace(-1, 1, Delta)
        Quant = np.array([])
        for i in time_compon[k][frame]:
            Quant = np.append(Quant,quantize(i,levels))
        frames_Quant.append(Quant)
    time_compon_Quant_stable.append(frames_Quant)
```

Step 2.3

```python
# Interpolation and Filtering with Adjustive Quantizer

output_signals = []
for k in range(M):
    frames = []
    for frame in range(1179):
        w = np.array([])
        for i in time_compon_Quant[k][frame][0:-1]:
            w = np.append(w,i)
            w = np.append(w,np.zeros(31))
        w = np.append(w,time_compon_Quant[k][frame][-1])
        frames.append(np.convolve(w, G_Filters[k],'full'))
    output_signals.append(frames)
```

```python
# Summation with Adjustive Quantizer

final_frames = output_signals[0]
k = 1

while k < 32:
    for frame in range(1179):
        final_frames[frame] = final_frames[frame] + output_signals[k][frame]  
    k = k + 1
```

```python
# Overlap add with Adjustive Quantizer

music_final = final_frames[0][:512]
for frame in range(1178):
    music_final = np.append(music_final, final_frames[frame][512:] + final_frames[frame + 1][:96])
    music_final = np.append(music_final, final_frames[frame + 1][96:512])
    
print(np.size(music_final) - np.size(music)) # Additional Zeros added initially

final = music_final*MAX

sf.write('adjastive_Quant.wav', final, 44100)
```

```python
# Interpolation and Filtering with Non adjastive Quantizer

output_signals_bad = []
for k in range(M):
    frames = []
    for frame in range(1179):
        w = np.array([])
        for i in time_compon_Quant_stable[k][frame][0:-1]:
            w = np.append(w,i)
            w = np.append(w,np.zeros(31))
        w = np.append(w,time_compon_Quant_stable[k][frame][-1])
        frames.append(np.convolve(w, G_Filters[k],'full'))
    output_signals_bad.append(frames)
```

```python
# Summation with Non adjastive Quantizer

final_frames_bad = output_signals_bad[0]
k = 1

while k < 32:
    for frame in range(1179):
        final_frames_bad[frame] = final_frames_bad[frame] + output_signals_bad[k][frame]  
    k = k + 1
    
```

```python
# Overlap add with Non adjastive Quantizer

music_final_bad = final_frames_bad[0][:512]
for frame in range(1178):
    music_final_bad = np.append(music_final_bad, final_frames_bad[frame][512:] + final_frames_bad[frame + 1][:96])
    music_final_bad = np.append(music_final_bad, final_frames_bad[frame + 1][96:512])
    
print(np.size(music_final_bad) - np.size(music)) # Additional Zeros added initially

final_bad = music_final_bad*MAX

sf.write('Non_adjastive_Quant.wav', final_bad, 44100)
```

```python
#Plotting first 512 specimens of output signals from diff Quantizers along with the input signal

%matplotlib inline
n = np.arange(0,512,1)

plt.plot(n,final[:512])

plt.plot(n,final_bad[:512])

plt.plot(n,music_mono[:512])
plt.xlabel('Specimens')
plt.title('Signals Comparison')    
```

```python
# Plot above helps calculate the delay of output signal

n0 = abs(np.argmax(music_mono[:256]) - np.argmax(final[:256]))
n0_bad = abs(np.argmax(music_mono[:256]) - np.argmax(final_bad[:256]))
```

```python
# Cutting the delay

final_cut = final[n0:]
final_bad_cut = final_bad[n0_bad:]

# and plotting first and last 512 specimens again
plt.figure(1)
n = np.arange(0,512,1)

plt.plot(n,final_cut[:512])

plt.plot(n,final_bad_cut[:512])

plt.plot(n,music_mono[:512])
plt.xlabel('First 512 pecimens') 
plt.title('Signals Comparison after cutting the delay')    

plt.figure(2)
plt.plot(n,final_cut[-512:])

plt.plot(n,final_bad_cut[-512:])

plt.plot(n,music_mono[-512:])
plt.xlabel('Last 512 specimens')
plt.title('Signals Comparison after cutting the delay')    
```

```python
# Plots above indicate that i should adittionally cut signals' end specimens so that they have length equal to the initial

diff = abs(np.size(music_mono) - np.size(final_cut))
diff_bad = abs(np.size(music_mono) - np.size(final_bad_cut))

Final_cut = final_cut[:-diff]
Final_bad_cut = final_bad_cut[:-diff_bad]
```

```python
# I will use the derived signals to calculate Quantization MSE

Error = Final_cut - music_mono
Error_bad = Final_bad_cut - music_mono

n = np.arange(0,512,1)

plt.plot(n,Error[:512])

plt.plot(n,Error_bad[:512])
plt.xlabel('First 512 specimens')
plt.title('Error Comparison')    

MSE = np.mean(np.square(Error))
MSE_bad = np.mean(np.square(Error_bad))

print('Mean square error of Adjustive Quantizer is: ' + str(MSE))
print('Mean square error of Non Adjustive Quantizer is: ' + str(MSE_bad))
```

```python
# Compression percentage

Total_Bits_per_frame_adjusted = sum([sum(Bits[i]*512) for i in range(32)]) # Different num of bits per filter per frame

Total_Bits_per_frame_initial = 16*np.size(music_mono)*32 # Bits multiplied by the num of filters to be comparable with the num of bits above

Total_Bits_per_frame_stable = 8*np.size(final_bad)*32

Compress = (Total_Bits_per_frame_initial - Total_Bits_per_frame_adjusted)/Total_Bits_per_frame_initial
Compress_bad = (Total_Bits_per_frame_initial - Total_Bits_per_frame_stable)/Total_Bits_per_frame_initial

print('Compression percentage with Adjustive quantizer is: ' + str(int(Compress*100)) + '%')
print('Compression percentage with Non Adjustive quantizer is: ' + str(int(Compress_bad*100)) + '%' )
```
